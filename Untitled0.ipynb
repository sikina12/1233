```python
import IPython.display as display

def login_app():
    username_correct = "usuario"
    password_correct = "senha123"

    print("Bem-vindo ao aplicativo de login!")
    username_input = input("Digite seu nome de usuário: ")
    password_input = input("Digite sua senha: ")

    if username_input == username_correct and password_input == password_correct:
        html_output = """
        <div style="background-color: lightgreen; padding: 20px; text-align: center; font-size: 24px; color: black;">
            Logado com sucesso!
        </div>
        """
        display.display(display.HTML(html_output))
    else:
        html_output = """
        <div style="background-color: lightcoral; padding: 20px; text-align: center; font-size: 24px; color: black;">
            Senha errada ou usuário incorreto!
        </div>
        """
        display.display(display.HTML(html_output))

# Para usar o aplicativo, chame a função:
login_app()
```
